```python
#Converting income into 0 & 1 
data.income.replace(('<=50K', '>50K'),(0,1), inplace=True)
```

```python
#Splitting into numerical & categorical feature
num_col = [col for col in data.columns if data[col].dtype != 'O']
cat_col = [col for col in data.columns if data[col].dtype == 'O']
```

```python
#Handling Outliers
```

```python
num_df=data[num_col]
cat_df=data[cat_col]
```

```python
# Checking for outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=num_df)
plt.title("Before Removing Outliers")
```

```python
#Creating function to handle outliers 
def outliers_imputation(num_df,col):
    IQR=num_df[col].quantile(0.75)-num_df[col].quantile(0.25)
    lower_fence=num_df[col].quantile(0.25)-(IQR*1.5)
    upper_fence=num_df[col].quantile(0.75)+(IQR*1.5)
    num_df.loc[num_df[col]<=lower_fence,col]=lower_fence
    num_df.loc[num_df[col]>=upper_fence,col]=upper_fence
```

```python
col=num_df.columns
```

```python
for j in col:
    outliers_imputation(num_df,j)
```

```python
# Box plot for all the features after handling outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=num_df)
plt.title("After Removing Outliers")
```
